```python
eff_0 = pd.read_csv('/ssd2/ming/2024COVID/output/cnn_10_total_pred_df.csv')
eff_0 = eff_0[['path','total_pred']]
new_path =[]
for i in eff_0.path.values:
    path_ = i.split('/')[-1]
    new_path.append(path_)
eff_0['path'] = new_path
eff_0 = eff_0.sort_values('path').rename(columns={"path":"name","total_pred":"pred"})
file="Submission1"
use_df=eff_0
folder_path="/ssd2/ming/2024COVID/submission"+"/"+file
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
use_df["name"]=use_df["name"].apply(lambda x : x.split("/")[-1])
covid_ct=use_df[use_df["pred"]>0.5].name.values
non_covid_ct=use_df[use_df["pred"]<0.5].name.values

covid_df=pd.DataFrame(covid_ct,columns=["ct_name"])
non_covid_df=pd.DataFrame(non_covid_ct,columns=["ct_name"])
covid_df.sort_values(by=['ct_name'],inplace=True)
non_covid_df.sort_values(by=['ct_name'],inplace=True)

covid_df.to_csv(f"{folder_path}/covid.csv",header=False,index=False)
non_covid_df.to_csv(f"{folder_path}/non-covid.csv",header=False,index=False)
print("covid:",len(covid_df) , "  non_covid:",len(non_covid_df))
    
```

```python
merged_df = eff_1.merge(eff_2, on="path", suffixes=('_1', '_2')).merge(eff_3, on="path")
merged_df = merged_df.rename(columns={"total_pred": "total_pred_3"})
merged_df['avg'] = merged_df[['total_pred_1', 'total_pred_2', 'total_pred_3']].mean(axis=1)

merged_df.to_csv("/ssd2/ming/2024COVID/submission/merge.csv",header=False,index=False)

merged_df = merged_df[['path','avg']]
new_path =[]
for i in merged_df.path.values:
    path_ = i.split('/')[-1]
    new_path.append(path_)
merged_df['path'] = new_path
merged_df = merged_df.sort_values('path').rename(columns={"path":"path","avg":"avg"})
file="Submission5"
use_df=merged_df
folder_path="/ssd2/ming/2024COVID/submission"+"/"+file
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
use_df["path"]=use_df["path"].apply(lambda x : x.split("/")[-1])
covid_ct=use_df[use_df["avg"]>0.50]
non_covid_ct=use_df[use_df["avg"]<0.50]

covid_df=pd.DataFrame(covid_ct,columns=["path"])
non_covid_df=pd.DataFrame(non_covid_ct,columns=["path"])
covid_df.sort_values(by=['path'],inplace=True)
non_covid_df.sort_values(by=['path'],inplace=True)

covid_df.to_csv(f"{folder_path}/covid.csv",header=False,index=False)
non_covid_df.to_csv(f"{folder_path}/non-covid.csv",header=False,index=False)
print("covid:",len(covid_df) , "  non_covid:",len(non_covid_df))
```
